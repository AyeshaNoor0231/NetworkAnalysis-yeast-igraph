```r
# Calculate the degree distribution
degree_distribution <- degree(yeast)
cat("Average degree:", mean(degree_distribution), "\n")

# Plot the degree distribution
hist(degree_distribution, main = "Degree Distribution", xlab = "Degree", ylab = "Frequency")
```


In an **adjacency matrix**, each element represents the presence or absence of an edge between a pair of vertices (nodes) in a graph. A 1 in the matrix typically represents an edge between the corresponding vertices. In network, the average degree is approximately 9.059992, which means, on average, **each node is connected to about 9 other nodes in the network**.

The **histogram** shows that the majority of nodes have degrees between 0 and 20, with a peak frequency of around 2000. As the degree increases beyond 20, the frequency decreases significantly, indicating that **fewer nodes have higher degrees in the networ**k.

```r
degree_centrality <- degree(yeast, mode = "all")
betweenness_centrality <- betweenness(yeast)
closeness_centrality <- closeness(yeast)

# Print summary statistics of centrality measures
print("Degree Centrality Summary:")
print(summary(degree_centrality))

print("\nBetweenness Centrality Summary:")
print(summary(betweenness_centrality))

print("\nCloseness Centrality Summary:")
print(summary(closeness_centrality))
```

```r
transitivity <- transitivity(yeast)
cat("Transitivity (Clustering Coefficient):", transitivity, "\n")
```

```r
# Plot degree centrality distribution
hist(degree_centrality, main = "Degree Centrality Distribution", xlab = "Degree Centrality", ylab = "Frequency")

# Plot betweenness centrality distribution
hist(betweenness_centrality, main = "Betweenness Centrality Distribution", xlab = "Betweenness Centrality", ylab = "Frequency")

# Plot closeness centrality distribution
hist(closeness_centrality, main = "Closeness Centrality Distribution", xlab = "Closeness Centrality", ylab = "Frequency")



```

**Betweenness Centrality:**
* The minimum betweenness of 0.0 indicates that some nodes are not part of any shortest path between other nodes.

* The mean betweenness of 4412.4 indicates the average number of shortest paths that pass through each node, highlighting nodes that act as bridges or mediators in the network.

**Closeness Centrality:**
* The minimum closeness of 0.0000411 suggests there are nodes that are relatively isolated from the rest of the network.
* The median closeness of 0.0000874 indicates the median reciprocal distance of each node to all other nodes, providing a measure of how close a node is to all other nodes on average.
* The mean closeness of 0.0588670 indicates the average closeness centrality across all nodes, reflecting the overall efficiency of information flow in the network.

* 1e+05, the frequency drops near to 0 suggests that there are very few nodes with high betweenness centrality values, indicating that only a small number of nodes act as critical intermediaries in connecting different parts of the network.

```r
# Select a random subset of vertices
random_vertices <- sample(V(yeast), 500)

# Extract the subgraph induced by these vertices
subgraph <- induced_subgraph(yeast, random_vertices)

# Plot the subgraph with attributes
plot(subgraph,
     vertex.label = V(subgraph)$Class,   # Displaying Class attribute as vertex label
     vertex.size = 3,                   # Setting vertex size
     edge.arrow.size = 0.5,             # Setting edge arrow size
     main = "Subgraph of Yeast Protein Interaction Network",  # Adding a title
     vertex.label.cex = 0.8)            # Adjusting vertex label size


```

```r
# Get the degrees of all vertices
degrees <- degree(subgraph)

# Set a threshold for the minimum degree
threshold <- quantile(degrees, 0.8)  # Selecting nodes with degrees above the 80th percentile

# Get the vertices with degrees above the threshold
high_degree_vertices <- which(degrees > threshold)

# Extract the subgraph induced by the high degree vertices
high_degree_subgraph <- induced_subgraph(subgraph, high_degree_vertices)

# Plot the subgraph with attributes
plot(high_degree_subgraph,
     vertex.label = V(high_degree_subgraph)$Class,   # Displaying Class attribute as vertex label
     vertex.size = 3,                   # Setting vertex size
     edge.arrow.size = 0.5,             # Setting edge arrow size
     main = "Subgraph with High Degree Vertices",  # Adding a title
     vertex.label.cex = 0.8)            # Adjusting vertex label size
```

```r
# Get the degrees of all vertices
degrees <- degree(subgraph)

# Sort the vertices based on their degrees in descending order
sorted_vertices <- order(degrees, decreasing = TRUE)

# Set the number of top nodes to select
top_nodes <- 20  # You can adjust this number based on your preference

# Get the top nodes with the highest degrees
top_degree_vertices <- sorted_vertices[1:top_nodes]

# Extract the subgraph induced by the top degree vertices
top_degree_subgraph <- induced_subgraph(subgraph, top_degree_vertices)

# Plot the subgraph with attributes
plot(top_degree_subgraph,
     vertex.label = V(top_degree_subgraph)$Class,   # Displaying Class attribute as vertex label
     vertex.size = 20,                   # Setting vertex size
     edge.arrow.size = 1,             # Setting edge arrow size
     main = "Subgraph with Top Degree Vertices",  # Adding a title
     vertex.label.cex = 1)            # Adjusting vertex label size
```

```r
node_to_plot <- "YDL203C"

# Get the index of the chosen node
node_index <- which(V(yeast)$name == node_to_plot)
searched_color <- "red"  # Color for the searched node
neighbor_color <- "blue"
# Get the neighbors of the chosen node
neighbors <- neighbors(yeast, node_index)

# Create a subgraph containing only the edges of the chosen node and its neighbors
subgraph <- induced_subgraph(yeast, c(node_index, neighbors))
# Define size for the nodes
searched_size <- 15  # Size for the searched node
neighbor_size <- 3  # Size for the neighboring nodes

# Plot the subgraph with different sizes for the searched node and its neighbors
plot(subgraph, vertex.label = NA, edge.arrow.size = 0.5,
     vertex.size = ifelse(V(subgraph)$name == node_to_plot, searched_size, neighbor_size),
     vertex.color = ifelse(V(subgraph)$name == node_to_plot, searched_color, neighbor_color))
```

```r
adj_matrix <- as_adjacency_matrix(subgraph, sparse = FALSE)

# Compute the layout for the subgraph
layout <- igraph::layout_with_fr(graph = subgraph)

# Plot the subgraph with different sizes for the searched node and its neighbors
plot(subgraph, layout = layout,
     vertex.size = ifelse(V(subgraph)$name == node_to_plot, searched_size, neighbor_size),
     vertex.color = ifelse(V(subgraph)$name == node_to_plot, searched_color, neighbor_color),
     edge.arrow.size = 0.5, vertex.label.dist = 1.5)

# Add vertex labels
text(layout[, 1], layout[, 2], labels = V(subgraph)$name, cex = 0.8, pos = 1)
```
